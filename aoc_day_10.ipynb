```python
# get list of adapters
li = list(df.adapter)

# set current addapter to 0
current_adapter = 0

# set list of diffs to empty list plus one three for final connection
diffs = [3]

# begin loop
running = True
while running:
    
    # set next adapter to minimun value in adapter list
    next_adapter = min(li)
    
    # remove current min adapter from list of adapters
    li.remove(min(li))
    
    # get differince in current and next adapter values
    diff = next_adapter - current_adapter
   
    # append diff to list of diffs
    diffs.append(diff)
    
    # change adaptor to next addapter
    current_adapter = next_adapter
    
    # break loop if li is empty
    if len(li) == 0:
        
        break
    
# get count of charges
num_one_diff = diffs.count(1)
num_three_diff = diffs.count(3)

# get number of one diffs times number of three diffs
print(num_one_diff," X ",num_three_diff," = ",num_one_diff*num_three_diff) 
```

```python
joltages = list(df.adapter)
joltages.sort()
```

```python
from functools import lru_cache

@lru_cache(None)
def dp(index):
    
    # at end of joltage list return 1
    if index == len(joltages) - 1:
        return 1

    # set total and nextIndex
    total = 0
    nextIndex = index + 1
    
    # while next index is not out of index and the differince in jolts between index and nextIndex is less than 3
    while nextIndex < len(joltages) and joltages[nextIndex] - joltages[index] <= 3:
        
        # call dp and add its value to total
        total += dp(nextIndex)
        
        # add one to nextIndex
        nextIndex += 1

    return total

dp(0)
```
